```python
import pandas as pd
```

```python
deci_macd = pd.read_excel('decimalodd_res/macd.xlsx', index_col=0)
deci_macd
```

```python
deci_macd_profit = deci_macd['Rate of Return'][deci_macd['Rate of Return'] > 0].sum()/len(deci_macd)
deci_macd_loss = deci_macd['Rate of Return'][deci_macd['Rate of Return'] < 0].sum()/len(deci_macd)
```

```python
print(deci_macd_profit, deci_macd_loss)
```

```python
wr_macd = pd.read_excel('winrate_result/macd.xlsx', index_col=0)
wr_macd
```

```python
zero_macd = wr_macd.columns[(wr_macd == 0).all()]
wr_macd['macd_average'] = wr_macd.drop(columns=zero_macd).mean(axis=1)
wr_macd
```

```python
wr_bb = pd.read_excel('winrate_result/bb.xlsx', index_col=0)
zero_bb = wr_bb.columns[(wr_bb == 0).all()]
wr_bb['bb_average'] = wr_bb.drop(columns=zero_bb).mean(axis=1)
wr_bb
```

```python
wr_obv = pd.read_excel('winrate_result/obv.xlsx', index_col=0)
zero_obv = wr_obv.columns[(wr_obv == 0).all()]
wr_obv['obv_average'] = wr_obv.drop(columns=zero_obv).mean(axis=1)
wr_obv
```
